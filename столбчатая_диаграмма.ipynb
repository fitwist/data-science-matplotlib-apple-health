## Влияние регул на тренировки

```python
import datetime
```

```python
mf = pd.read_csv('/Users/helenkapatsa/Repositories/applehealthdata/kea/MenstrualFlow.csv')
mf.creationDate[1]
```

```python
mfTimestampStr = []
mfTimestamp = []

for i in range(len(mf.creationDate)):
    mfTimestampStr.append(mf.creationDate[i])
    
for i in range(len(mfTimestampStr)):
    mfTimestamp.append(datetime.datetime.strptime(mfTimestampStr[i], '%Y-%m-%d %H:%M:%S +%f').date())
    print(mfTimestamp[i])
```

```python
elTimestampStr = []
elTimestamp = []

for i in range(len(el.creationDate)):
    elTimestampStr.append(el.creationDate[i])
    
for i in range(len(elTimestampStr)):
    elTimestamp.append(datetime.datetime.strptime(elTimestampStr[i], '%Y-%m-%d %H:%M:%S +%f').date())
    print(elTimestamp[i])
```

```python
plt.bar(x = elTimestamp,
        color = 'b',
       height = elTimestamp)

plt.bar(x = mfTimestamp,
        color = 'c',
       height = mfTimestamp)

plt.yticks([])
```
